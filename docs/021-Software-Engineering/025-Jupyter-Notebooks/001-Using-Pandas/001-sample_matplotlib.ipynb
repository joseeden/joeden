Next, create the code. Import `pyplot`. The graph below is **linear**

````python
import matplotlib.pyplot as plt
year = [1970, 1980, 1990, 2000, 2010, 2020]
pop = [3.7, 4.4, 5.3, 6.1, 6.9, 7.8]

plt.plot(year, pop)
plt.show()
````

To create a scatter plot:

````python
import matplotlib.pyplot as plt
year = [1970, 1980, 1990, 2000, 2010, 2020]
pop = [3.7, 4.4, 5.3, 6.1, 6.9, 7.8]

plt.scatter(year, pop)
plt.show
````

# Histograms

````python
import matplotlib.pyplot as plt
data = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(data, bins=6)
plt.show()
````

Using another example:

````python
import matplotlib.pyplot as plt

years = [1950, 2000, 2050, 2100]
pop = [2.5, 6.1, 9.0, 10.5]

plt.plot(years, pop)
plt.show()
````

Adding X and Y labels, as well the plot title.

````python
import matplotlib.pyplot as plt

years = [1950, 2000, 2050, 2100]
pop = [2.5, 6.1, 9.0, 10.5]

plt.plot(years, pop)
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.title('World Population Projections')
plt.show()
````

````python
import matplotlib.pyplot as plt

years = [1950, 2000, 2050, 2100]
pop = [2.5, 6.1, 9.0, 10.5]

plt.plot(years, pop)
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8, 10], ['0', '2B', '4B', '6B', '8B', '10B'])
plt.show()
````

Adding ticks:

````python
import matplotlib.pyplot as plt

years = [1950, 2000, 2050, 2100]
pop = [2.5, 6.1, 9.0, 10.5]

years = [1800, 1850, 1900] + years
pop = [1.0, 1.3, 1.6] + pop

plt.plot(years, pop)
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8, 10], ['0', '2B', '4B', '6B', '8B', '10B'])
plt.show()
````
