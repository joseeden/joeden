## Fetching Weather Data using OpenWeatherMap API

This notebook shows how to get current weather data for multiple cities using the OpenWeatherMap API. The data will be fetched, combined, and displayed in a readable JSON format.

## Create API Helper and Fetch Weather Data

Set up the API key and a helper function to fetch weather data for any city.

```python
import requests
import json

def get_weather(city):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return {
            "city": city,
            "temp_c": data["main"]["temp"],
            "condition": data["weather"][0]["main"],
            "humidity": data["main"]["humidity"]
        }
    else:
        return {"city": city, "error": "Could not fetch data"}
```

Next, fetch the weather data for multiple cities. 

The code below loops through each city in the `cities` list and then stores results in a dictionary under `"weather"`.

For example, London and Paris.

```python
cities = ["London", "Paris"]
weather_data = [get_weather(city) for city in cities]
payload = {"weather": weather_data}

print("Current weather report:")
print(json.dumps(payload, indent=2))
```

## Make it Interactive

Add more cities interactively:

```python
# Ask user for comma-separated city names
user_input = input("Enter cities separated by commas: ")
cities = [city.strip() for city in user_input.split(",")]

# Fetch weather
weather_data = [get_weather(city) for city in cities]
payload = {"weather": weather_data}

print(json.dumps(payload, indent=2))
```
